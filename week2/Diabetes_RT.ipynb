```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

# 1. 데이터 로드
path = '/content/drive/MyDrive/Colab Notebooks/diabetes.csv'
df = pd.read_csv(path)

# 2. 데이터 분리 (Target: BMI)
# BMI를 예측하므로 X에서 Outcome과 BMI를 모두 제외합니다.
y = df['BMI']
X = df.drop(['Outcome', 'BMI'], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# 3. 모델 생성 및 학습 (회귀 트리)
reg_rt = DecisionTreeRegressor(random_state=0)
reg_rt.fit(X_train, y_train)

# 4. 예측 및 평가
pred_rt = reg_rt.predict(X_test)
mse = mean_squared_error(y_test, pred_rt)

print("--- Regression Tree (Regression) ---")
print("Target: BMI")
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", np.sqrt(mse))
```
