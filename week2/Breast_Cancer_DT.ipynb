```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

# 1. 파일 경로 설정 및 로드
path = '/content/drive/MyDrive/Colab Notebooks/breast_cancer.csv'
df = pd.read_csv(path, index_col=0)

# 2. 데이터 분리
y = df['worst concavity']
X = df.drop(['worst concavity', 'label'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# 3. DT 모델 생성 및 학습
model_dt = DecisionTreeRegressor(random_state=0)
model_dt.fit(X_train, y_train)

# 4. 예측 및 평가
pred_dt = model_dt.predict(X_test)
print("\n--- Decision Tree (DT) ---")
print('Mean Squared Error (MSE):', mean_squared_error(y_test, pred_dt))
```
